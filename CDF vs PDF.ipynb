```python
import pandas as pd
import numpy as np
import scipy.stats as si
```

```python
def NORMSINV(x):
    NORMSINV = si.norm.ppf(x,0.0,1.0)
    return(NORMSINV)
RAND = np.random.standard_normal()
```

```python
# XYZ
mean1 = 0.26 # Expected return (mean)
vol1 = 0.60 # Volatility (Std Dev)
# IBM
mean2 = 0.13 # Expected return (mean)
vol2 = 0.30 # Volatility (Std Dev)
# Correlation
corr = 0.2 # Correlation (rho)
```

```python
Return = np.linspace(-1.5,1.5,31)
print(Return)
```

# Cumulative Distribution Functions (CDF)

```python
XYZ_CDF = []
for i in range(0,len(Return)):
    XYZ_CDF.append(si.norm.cdf(Return[i],mean1,vol1))
    
IBM_CDF = []
for i in range(0,len(Return)):
    IBM_CDF.append(si.norm.cdf(Return[i],mean2,vol2))
```

```python
cdf = pd.DataFrame({"Return": Return, "CXYZ": XYZ_CDF, "CIBM": IBM_CDF})
cdf
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))
IBM_CDF = plt.plot(cdf.Return, cdf.CIBM, label='IBM', color = 'crimson', linewidth = 7.0)
XYZ_CDF = plt.plot(cdf.Return, cdf.CXYZ, label='XYZ', color = 'blue', linewidth = 7.0)
plt.title('Cumulative Density Function (Cdf)')
plt.xlabel('Return')
plt.ylabel('IBM vs. XYZ')
plt.legend()
plt.show()
```

# Probability Density Function (PDF)

```python
XYZ_PDF = []
for i in range(0,len(Return)):
    XYZ_PDF.append(si.norm.pdf(Return[i],mean1,vol1))
    
IBM_PDF = []
for i in range(0,len(Return)):
    IBM_PDF.append(si.norm.pdf(Return[i],mean2,vol2))
```

```python
pdf = pd.DataFrame({"Return": Return, "PXYZ": XYZ_PDF, "PIBM": IBM_PDF})
pdf
```

```python
import matplotlib.pyplot as plt
# Cumulative counts:
fig = plt.figure(figsize=(10,10))
IBM_PDF = plt.plot(pdf.Return, pdf.PIBM, label='IBM', color = 'crimson', linewidth = 7.0)
XYZ_PDF = plt.plot(pdf.Return, pdf.PXYZ, label='XYZ', color = 'blue', linewidth = 7.0)
plt.title('Probability Density Function (pdf)')
plt.xlabel('Return')
plt.ylabel('IBM vs. XYZ')
plt.legend()
plt.show()
```
