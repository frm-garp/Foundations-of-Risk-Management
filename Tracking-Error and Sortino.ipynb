```python
import pandas as pd
import numpy as np
import scipy.stats as si
```

```python
def NORMSINV(x):
    NORMSINV = si.norm.ppf(x,0.0,1.0)
    return(NORMSINV)
RAND = np.random.standard_normal()
```

```python
RAND
```

```python
RF = 0.04 # Risk free Rate
MAR = 0.05 # Minimum Acceptable Return (MAR)
```

```python
# Bench (B)
meanB = 0.04 # Exp Return
volB = 0.1 # Std Deviation
```

```python
# Portfolio (P)
meanP = 0.08 # Exp Return
volP = 0.1 # Std Deviation
```

```python
Corr = 0.5 # Correlation
cov = volP*volB*Corr # Covariance
```

```python
Period = ['t-1', 't-2', 't-3', 't-4', 't-5', 't-6', 't-7', 't-8', 't-9', 't-10',
         't-11', 't-12', 't-13', 't-14', 't-15', 't-16', 't-17', 't-18', 't-19', 't-20',
         't-21', 't-22', 't-23', 't-24', 't-25', 't-26', 't-27', 't-28', 't-29', 't-30']
Period
```

```python
B_RAND = np.random.rand(len(Period))
B_RAND
```

```python
B_N = []
for i in range(0,len(Period)):
    B_N.append(NORMSINV(B_RAND[i]))
B_N   
```

```python
P_RAND = np.random.rand(len(Period))
P_RAND
```

```python
P_N = []
for i in range(0,len(Period)):
    P_N.append(Corr*B_RAND[i]+np.sqrt((1-Corr**2))*NORMSINV(P_RAND[i]))
P_N
```

```python
B = []
for i in range(0,len(Period)):
    B.append(meanB+volB*B_N[i])
B   
```

```python
P = []
for i in range(0,len(Period)):
    P.append(meanP+volP*P_N[i])
P   
```

```python
P_minus_B = []
for i in range(0,len(Period)):
    P_minus_B.append(P[i]-B[i])
P_minus_B  
```

```python
P_minus_MAR = []
for i in range(0,len(Period)):
    if P[i] < MAR:
        P_minus_MAR.append(P[i]-MAR)
    else:
        P_minus_MAR.append(0)
P_minus_MAR
```

```python
P_minus_MAR_s = []
for i in range(0,len(Period)):
    if P[i] < MAR:
        P_minus_MAR_s.append((P[i]-MAR)**2)
    else:
        P_minus_MAR_s.append(0)
P_minus_MAR_s
```

```python
df = pd.DataFrame({"Period": Period,
                   "B_N(0,1)":B_N,
                   "P_N(0,1)":P_N,
                   "Benchmark (B)":B,
                   "Portfolio (P)":P,
                   "Active Return (P-B)":P_minus_B,
                   "(P-MAR)":P_minus_MAR,
                   "Squared (P-MAR)":P_minus_MAR_s})
df
```

```python
Avg_P = np.mean(P)
```

```python
df["P-Avg_P"] = 0.000
for i in range(len(Period)):
    df["P-Avg_P"][i] = (df["Portfolio (P)"][i]-Avg_P)**2
```

```python
df
```

```python
Count = len(df["Portfolio (P)"])
print("The Count of Returns is:", Count)
```

```python
P_MAR = np.sum(df["Squared (P-MAR)"])
print("The Sum of Squared Differences (P-MAR) is:", "{:.2%}".format(P_MAR))
```

```python
DD = np.sqrt(P_MAR/Count)
print("The Downside deviation is:", "{:.2%}".format(DD))
```

```python
from statistics import stdev
sd = stdev(df["Portfolio (P)"])
print("The Standard Deviation is:", "{:.2%}".format(sd))
```

```python
wea = np.sqrt(volP**2+volB**2-2*volB*volP*Corr)
print("The Tracking Error (ex ante) is:", "{:.2%}".format(wea))
```

```python
wep = stdev(df["Active Return (P-B)"])
print("The Tracking Error (ex post) is:", "{:.2%}".format(wep))
```

```python
srea = (meanP-RF)/volP
print("The Sharpe ratio (ex ante) is:", round((srea),2))
```

```python
srep = (np.mean(df["Portfolio (P)"])-RF)/sd
print("The Sharpe ratio (ex post) is:", round((srep),2))
```

```python
sortino = (np.mean(df["Portfolio (P)"])-MAR)/DD
print("The Sortino with realized (ex post) return is:", round((sortino),2))
```
